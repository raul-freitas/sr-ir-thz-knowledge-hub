```python
import folium

# Create a map centered on São Paulo
m = folium.Map(location=[51.49219264029158, 0.006349814933169349], zoom_start=1)

# HTML content with a clickable link (opens in new tab)
IMB = """
<b>LNLS</b><br>
<a href="https://lnls.cnpem.br/en/" target="_blank">
Brazilian Synchrotron Light Laboratory - LNLS<br>
</a>
<br>
<a href="https://lnls.cnpem.br/facilities/imbuia-en/" target="_blank">
IMBUIA beamline<br>
</a>
Status: Operation at 200 mA <br>
SR nano-FTIR, s-SNOM imaging <br>
500-5000 cm⁻¹ (62-620 meV) <br>
<br>
<a href="https://lnls.cnpem.br/facilities/imbuia-en/" target="_blank">
TATU beamline<br>
</a>
Status: Construction <br>
SR THz nanospectroscopy, s-SNOM THz imaging <br>
100-500 cm⁻¹ (12-62 meV, 3-15 THz) <br>
<br>
<a href="https://sauonline.cnpem.br/duo/" target="_blank">
Submission Portal<br>
</a>
"""

SMIS = """
<b>SOLEIL</b><br>
<a href="https://www.synchrotron-soleil.fr/en" target="_blank">
Synchrotron SOLEIL<br>
</a>
<br>
<a href="https://www.synchrotron-soleil.fr/en/beamlines/smis" target="_blank">
SMIS beamline<br>
</a>
Status: Operation at 500 mA <br>
SR nano-FTIR, s-SNOM imaging, SR micro-FTIR <br>
500-5000 cm⁻¹ (62-620 meV) <br>
<br>
<a href="https://sun.synchrotron-soleil.fr/sunset/bridge/sunset/" target="_blank">
Submission Portal<br>
</a>
"""

BL24 = """
<b>ALS</b><br>
<a href="https://als.lbl.gov" target="_blank">
Advanced Light Source - ALS<br>
</a>
<br>
<a href="https://als.lbl.gov/beamlines/2-4/" target="_blank">
Beamline 2.4<br>
</a>
Status: Operation at 500 mA <br>
SR nano-FTIR, s-SNOM imaging, SR micro-FTIR <br>
300-5000 cm⁻¹ (62-620 meV) <br>
<br>
<a href="https://alshub.als.lbl.gov" target="_blank">
Submission Portal<br>
</a>
"""


# Create Popup
popupIMB = folium.Popup(IMB, max_width=300)
popupSMIS = folium.Popup(SMIS, max_width=300)
popup24 = folium.Popup(BL24, max_width=300)


# Add marker
folium.Marker(
    [-22.80732642294741, -47.053537019691106],
    popup=popupIMB,
    tooltip="LNLS"
).add_to(m)

folium.Marker(
    [48.70995646615653, 2.1452810394732476],
    popup=popupSMIS,
    tooltip="SOLEIL"
).add_to(m)

folium.Marker(
    [37.87576469216587, -122.24867612758241],
    popup=popup24,
    tooltip="ALS"
).add_to(m)


# Display map
m
```
